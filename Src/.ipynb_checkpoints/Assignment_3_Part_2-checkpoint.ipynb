#### Define categorical and numerical variables

```python
categorical_vars = ["Survived", "Pclass", "Name", "Sex", "Ticket", "Cabin", "Embarked"]
numerical_vars = ["Age", "SibSp", "Parch", "Fare"]

# Print categorized variables
categorized_variables = {
    "Categorical_Variables": categorical_vars,
    "Numerical_Variables": numerical_vars
}
categorized_variables
```

#### Define valid ranges for each variable

```python
valid_ranges = {
    "Survived": (0, 1),   # as mentioned in the data description
    "SibSp": (0, 8),      # as mentioned in the data description
    "Parch": (0, 8),      # as mentioned in the data description
    "Pclass": (1, 3),     # as mentioned in the data description
    "Fare": (7, 800),     # as mentioned by BBC https://www.bbc.co.uk/bitesize/articles/zng8jty
    "Age": (0.1, 80.0)     # as mentioned in the data description
}
```

#### Function to count and remove out of range variables


```python
def check_out_of_range(data, valid_ranges):
    results = {}
    for var, (lower, upper) in valid_ranges.items():
        if var in data.columns:
            out_of_range = (data[var] < lower) | (data[var] > upper)
            results[var] = out_of_range.sum()
        else:
            results[var] = None  # Variable not found in the dataset
    return results

def remove_out_of_range(data, valid_ranges):
    df = data.copy()
    for var, (lower, upper) in valid_ranges.items():
        if var in df.columns:
            if not pd.api.types.is_numeric_dtype(df[var]):
                df[var] = pd.to_numeric(df[var], errors='coerce')
            mask = df[var].isna() | ((df[var] >= lower) & (df[var] <= upper))
            df = df[mask]
    return df

# Check outliers before cleaning
titanic_data_outliers_before = check_out_of_range(titanic_data, valid_ranges)

# Clean dataset
titanic_data = remove_out_of_range(titanic_data, valid_ranges)

# Check outliers after cleaning
titanic_data_outliers_after = check_out_of_range(titanic_data, valid_ranges)

# Summary of before and after
print("Train Data - out of range Before and After Cleaning:")
print("Before:", titanic_data_outliers_before)
print("After:", titanic_data_outliers_after)
```

#### plotting and handling missing values

Count missing values per column

```python
print(titanic_data.isna().sum())
```

Visualize missing data using seaborn's heatmap

```python
plt.figure(figsize=(10, 6))
sns.heatmap(titanic_data.isnull(), cbar=False, cmap='viridis')
plt.title("Missing Data Heatmap")
plt.show()
```

Calculate missing percentage for each column

```python
missing_percentage = titanic_data.isna().sum() / len(titanic_data) * 100
print(missing_percentage)
```

Impute missing Age values with the median

```python
titanic_data['Age'] = titanic_data['Age'].fillna(titanic_data['Age'].median())
```

Impute missing Embarked values with the mode (most common value)

```python
mode_embarked = titanic_data['Embarked'].mode()[0]
titanic_data['Embarked'] = titanic_data['Embarked'].fillna(mode_embarked)
```

Replace missing Cabin values with "Unknown"

```python
titanic_data['Cabin'] = titanic_data['Cabin'].fillna("Unknown")
```

#### Define and Test Hypotheses


Hypothesis 1: Survival Rate and Passenger Class
Question: Is the survival rate associated with the class of the passenger?

Analysis:

Group the data by Pclass and calculate the survival rate.

Use a bar plot to visualize the survival rate 

Use Count plot of survival by passenger class for visualization as wellssociation.

```python
# Survival rate by class
survival_by_class = titanic_data.groupby('Pclass')['Survived'].mean().reset_index()

survival_by_class
```

```python
# Plot
sns.barplot(x='Pclass', y='Survived', data=survival_by_class)
plt.title('Survival Rate by Passenger Class')
plt.show()
```

```python
# Plot: Count plot of survival by passenger class
sns.countplot(x='Pclass', hue='Survived', data=titanic_data, palette='Set1')
plt.title("Survival Count by Passenger Class")
plt.show()
```

It is very clear using the above graphs tht the survival rate is definitly associated with the class of the passenger.

#### Hypothesis 3: Survival Rate and Age
Question: Is the survival rate associated with the age of the passenger?

Analysis:

Create Age groups to compare survival rates across different age groups

Use a bar plot to visualize the survival rate by age group.

Plot the age distribution (histogram) for survivors vs. non-survivors

Box plot to see age spread for each survival group

```python
# Create age groups
titanic_data['AgeGroup'] = pd.cut(titanic_data['Age'], bins=[0, 12, 18, 35, 60, 100])
#Survival rate by age group
survival_by_age = titanic_data.groupby('AgeGroup', observed=False)['Survived'].mean().reset_index() #used observed parameter to get rid of warning
survival_by_age
```

```python
# Plot
sns.barplot(x='AgeGroup', y='Survived', data=survival_by_age)
plt.title('Survival Rate by Age Group')
plt.show()
```

```python
sns.histplot(data=titanic_data, x='Age', hue='Survived', bins=8, kde=True, palette='Set1')
plt.title("Age Distribution by Survival")
plt.ylabel("Frequency")
plt.show()
```

```python
sns.boxplot(x='Survived', y='Age', data=titanic_data)
plt.title("Age Distribution for Survived vs. Not Survived")
plt.xlabel("Survived (0 = No, 1 = Yes)")
plt.ylabel("Age")
plt.show()
```


Age does seem to have some some kind of relation with survival rate but it's not as clear as pclass and gender.
It appears the most likely age a person on titanic has for not surviving is between the age of 20-35
